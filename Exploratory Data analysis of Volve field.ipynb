```python
#function for resetting index for each well bore and converting date columnn to datatime format

def dataformat(df, wellbore_name):
    """Function for resetting index for each well bore and converting data column to datetime format 
    df = dataframe of field 
    wellbore name = wellbore name whose data we want"""
    df_well = df[df['NPD_WELL_BORE_NAME']==wellbore_name]
    df_well = df_well.reset_index()
    df_well = df_well.drop("index", axis = 1)
    df_well["DATEPRD"] = pd.to_datetime(df_well['DATEPRD'])
    return df_well
    
```
