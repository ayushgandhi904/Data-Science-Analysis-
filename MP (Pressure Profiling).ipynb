## Pressure Profiling 

```python
B = 1
re = float(input("Enter outer radius of reservoir(ft): "))
rw = float(input("Enter wellbore radius(ft): "))
pwf = float(input("Enter bottomhole pressure(psi): "))
h = float(input("Enter the net pay thickness(ft): "))
k = float(input("Enter average Reservoir Permeability (mD): "))
q = float(input("Enter flowrate (STB/day): "))
mu = float(input("Enter the oil viscosity (cp): "))
```

```python
r = np.linspace(rw, re, 500)
```

```python
pressure = []
for i in range(len(r)):
    p = pwf + (141.2*q*mu*B*(np.log(r[i]/rw))/(k/h))
    pressure.append(p)
```

```python
plt.figure(figsize = (8, 6))
plt.plot(r, pressure)
plt.xlabel("r (ft)")
plt.ylabel("P(r), psi")
plt.title("Reservoir pressure Profile")
plt.grid()
plt.style.use("classic")
```

### Dynamic Pressure Profile: Visualizing the effect of viscosity, flowrate, and permeability 

```python
def flowprofile(k, mu, q):
    re = 3000
    rw = 0.5
    r = np.linspace(rw, re, 500)
    pe = 4000
    B = 1
    h = 30
    
    p = pe - (141.2*q*mu*B*(np.log(re/r))/k/h)
    
    y_m = p[np.where(r == rw)]
    
    plt.plot(r, p, lw = 4)
    plt.axhline(y_m, lw = 1, color = "red")
    
    plt.ylim(0, 5000)
    plt.xlabel("r(ft)")
    plt.ylabel("P(r), psi")
    
    plt.title("Reservoir Pressure Profile")
    
    plt.grid()
    
    return r, p
```

```python
w = interactive(flowprofile, k = (200, 1000), mu = (10, 220), q = (100, 200))
```
