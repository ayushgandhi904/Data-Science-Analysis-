## 1. Klinkenberg's effect 

#### Klinkenberg effect : The observed difference in gas and water permeabilities is analyzed by the Klinkenberg effect. The effect is due to the slip flow of gas at pore walls which enhances gas flow when pore sizes are very small. It is necessary to at least measure the gas permeability at two-mean pressure. 

##### It is calculated by the absolute permeability when only one gase permeability (kg) of a core sample is made at pm. This non-linear equation can be solved iteratively by using the Newton-Rapshon iterative method. 

```python
k = float(input("Enter guess of absolute perm(mD):"))
pm = float(input("Enter the mean pressure(psi): "))
kg = float(input("Enter the gas perm(mD): "))
```

```python
while abs(6.9*(k**0.64) + pm*k - pm*kg) > 0.0001:
    k -= (6.9*(k**0.64) + pm*k - pm*kg)/(4.416*(k**(-0.36)) + pm)
```

```python
print(f"The value of absolute permeability after iteration or gas permeability at infinite mean is {k}")
```

```python
# plotting graph
```

```python
x = [0, 1/pm]
y = [k, kg]
```

```python
plt.figure(figsize = (8, 4))
plt.plot(x, y)
plt.style.use('classic')
plt.xlabel('1/mean pressure (psi^-1)')
plt.ylabel('perm')
plt.title('klinkenberg effect')
plt.grid()
```
