```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Volve_PF14.csv")
```

```python
df.head()
```

```python
df.drop("NPD_WELL_BORE_CODE", axis = "columns", inplace = True)
```

```python
df = df.set_index("DATEPRD")
```

```python
df
```

```python
df["BORE_OIL_VOL"].plot(figsize = (12, 4));
plt.title("Water production");
```

```python
df.describe()
```

```python
ax1 = df["BORE_OIL_VOL"].plot(figsize = (12, 4), label = "Oil Production", legend = True, color = "orange");
ax1.set_ylabel("Oil rates", color = "orange", size = 12);

ax2 = df["BORE_WAT_VOL"].plot(figsize = (12, 4), label = "Water Productiion", legend = True, color = "blue");
ax2.set_ylabel("Water rates", color = "blue", size = 12);

ax2 = ax1.twinx()

plt.title("Oil & Water Production Profiles of Volve Well  F-14", size = 15);
plt.grid()
```

```python
plt.figure(figsize = (12, 4))
sns.boxplot(df["BORE_OIL_VOL"], color = "orange")
sns.boxplot(df["BORE_WAT_VOL"], color = "blue")
```
